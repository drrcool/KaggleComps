# Exploration of the Forest Cover competition from Kaggle

This is a first pass at the data to visualize what's in it and how various things relate to each other.

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

%matplotlib inline
```

```python
# Read the data in
train = pd.read_csv("data/train.csv")
test = pd.read_csv('data/test.csv')
```

First order looking at the data says it's in pretty good shape.  All of the columns are either categorical and already split, and thus binary (such as Soil_Type).  Looking at train.describe(), it doesn't even look like there's much in terms of missing data. For now, I'm not going to even do any cleanup on the data. We can change that later.

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.scatter(train['Horizontal_Distance_To_Hydrology'], 
         train['Vertical_Distance_To_Hydrology'],
         marker='.')
plt.xlabel("Horiz. Dist to Water")
plt.ylabel("Vert. Dist to Water")
plt.ylim(-200, 400)
plt.xlim(0, 1400)
colors = ['red','orange','yellow','green','blue','cyan', 'indigo', 'violet']
for i in range(1,8):
    subset = train[train['Cover_Type']==i]
    plt.scatter(subset['Horizontal_Distance_To_Hydrology'], 
         subset['Vertical_Distance_To_Hydrology'],
         marker='.', color=colors[i-1])
```

```python
# Let's look at a giant Seaborn plot
useful_feats_to_plot = ['Elevation',
                        'Horizontal_Distance_To_Hydrology', 
                        'Vertical_Distance_To_Hydrology',
                        'Horizontal_Distance_To_Roadways', 
                        'Slope']

nplot = 1000 # to make it easier to see
subsample = train.ix[random.sample(range(len(train)), nplot),:]
sns.pairplot(subsample, 
             hue="Cover_Type", 
             vars=useful_feats_to_plot)
```
